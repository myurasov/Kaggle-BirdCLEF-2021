```python
def _rectify_primary_label_counts(df, max_items, min_items):
    
    res_df = df.copy()

    for l, c in (zip(primary_label_counts.index, primary_label_counts)):

        if c > max_items:
            
            # too much - drop samples over mean+1std
            print(f"too much: {l=} {c=}")
            to_drop = df[df['primary_label'] == l].sample(c - max_items)
            res_df = res_df.drop(to_drop.index)
            
        elif c < min_items:
            
            # too little - add samples to mean - std
            print(f"\ntoo little: {l=} {c=}")
            
            # repeat existing rows
            repeats = int(math.ceil(min_items / c))
            rows_to_repeat = df[df['primary_label'] == l]
            
            for i in range(repeats - 1):
                res_df = res_df.append(rows_to_repeat, ignore_index=True)
                
            
            # drop excessive items
            c = res_df[res_df['primary_label'] == l].shape[0]
            to_drop = res_df[res_df['primary_label'] == l].sample(c - min_items)

            assert to_drop.shape[0] == c - min_items
            print(f' need to drop {to_drop.shape[0]=} out of {c=}')

            c = res_df[res_df['primary_label'] == l].shape[0]
            print(f" before {c=}")

            res_df = res_df.drop(to_drop.index)

            c = res_df[res_df['primary_label'] == l].shape[0]
            print(f" after {c=}")

                
    return res_df
```

```python
primary_label_counts = df["primary_label"].value_counts()
mean, std = np.mean(primary_label_counts), np.std(primary_label_counts)
print(f"{mean=:.0f} {std=:.0f}")    
df = _rectify_primary_label_counts(df, int(mean + std), int(mean - std))
```

```python
species = df["primary_label"].value_counts()

fig = go.Figure(
    data=[go.Bar(y=species.values, x=species.index)],
    layout=go.Layout(margin=go.layout.Margin(l=0, r=0, b=10, t=50)),
)

fig.show()
```
