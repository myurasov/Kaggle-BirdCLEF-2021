```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_pickle("/app/_work/dataset.pickle")
```

```python
f = 1 / np.array(
    list(
        df["_primary_labels"].replace(
            dict(df["_primary_labels"].value_counts()),
        ),
    )
)

f /= np.max(f)
```

```python
df['_f'] = f
```

```python
plt.plot(f)
plt.figure()
plt.plot(sorted(f))
```

```python
plt.plot(df["_f"])
```
